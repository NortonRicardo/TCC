```python
# Função para criar e salvar o mapa de calor para um ano específico
def criar_mapa_ano(ano):
    
    # Definindo as datas de início e fim do ano
    inicio_ano = f"{ano}-01-01"
    fim_ano = f"{ano}-12-31"

    print("=======================================")
    print(f"mapa do ano {inicio_ano} ate {fim_ano}")

    # Conectar ao banco de dados
    with psycopg2.connect(f"host={host} port={port} user={user} password={password} dbname={dbname}") as conn:
        query = f"""
        SELECT data.inmet_weather_station_id as cdg_estacao,
               station.vl_latitude as latitude,
               station.vl_longitude as longitude,
               station.cidade as cidade,
               station.sg_estado as estado,
               COUNT(*) as total_registros_com_null
        FROM inmet_weather_data as data
        LEFT JOIN inmet_weather_stations as station ON station.cdg_estacao = data.inmet_weather_station_id
        WHERE data.dta_medicao BETWEEN '{inicio_ano}' AND '{fim_ano}'
        AND station.sg_estado in ('GO', 'DF')
        AND (
            data.umd_ins IS NULL OR
            data.tem_ins IS NULL OR
            data.pre_ins IS NULL OR
            data.rad_glo IS NULL OR
            data.pto_ins IS NULL OR
            data.chuva IS NULL OR
            data.ven_dir IS NULL OR
            data.ven_vel IS NULL OR
            station.vl_latitude IS NULL OR
            station.vl_longitude IS NULL OR
            station.cidade IS NULL OR
            station.sg_estado IS NULL
        )
        GROUP BY data.inmet_weather_station_id, station.vl_latitude, station.vl_longitude, station.cidade, station.sg_estado
        ORDER BY total_registros_com_null DESC;
        """

        print("Fazendo consulta ")
        conn = psycopg2.connect(conn_string)
        
        df = pd.read_sql_query(query, conn)

        conn.close()

        print("Gerando mapa")

    # Verificar se o DataFrame não está vazio
    if not df.empty:
        # Criar um mapa centrado nas coordenadas médias do DataFrame
        mapa_centro = [df['latitude'].mean(), df['longitude'].mean()]
        mapa = folium.Map(location=mapa_centro, zoom_start=5)

        # Preparar os dados para o HeatMap
        heat_data = [[row['latitude'], row['longitude'], row['total_registros_com_null']] for index, row in df.iterrows()]
        HeatMap(heat_data, radius=20).add_to(mapa)

        # Loop para adicionar marcadores com tooltips
        for index, row in df.iterrows():
            tooltip_text = f"CDG: {row['cdg_estacao']}<br>Cidade: {row['cidade']}"
            
            # folium.Marker(
            #     location=[row['latitude'], row['longitude']],
            #     tooltip=tooltip_text,
            #     icon=folium.Icon(color='red', icon='info-sign')
            # ).add_to(mapa)

            folium.CircleMarker(
                location=[row['latitude'], row['longitude']],
                radius=1,  # Tamanho do marcador
                color='blue',
                fill=True,
                fill_color='blue',
                fill_opacity=1,
                tooltip=tooltip_text  # Repetindo o tooltip para o marcador
            ).add_to(mapa)

        # Salvar o mapa em um arquivo HTML para visualização
        print("Salvando mapa")
        mapa.save(f'results/html/mapa_calor_dados_faltantes/dados_faltantes_{ano}.html')


    print("Finalizado")
    print("=======================================")
```

```python
# Iterar sobre os anos de 2000 a 2023 e criar um mapa para cada ano
for ano in range(2000, 2024):
    criar_mapa_ano(ano)
```
